```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
class DatasetAnalyzer:
    def __init__(self):
        self.dataset_path = None
        self.data = None
        self.features = None
        self.target = None
        self.model = None

    def load_dataset(self):
        mat = pd.read_csv("student-mat.csv")
        por = pd.read_csv("student-por.csv")

        self.data = pd.concat([mat, por])

    def explore_dataset(self):

        # Типо всякие графики, зависимости и тп
        # Анализ датасета
        print("Dataset Info:")
        print(self.data.info())

        print("Correlation Matrix:")
        correlation_matrix = self.data.corr()
        print(correlation_matrix)

    def prepare_data(self, test_size=0.2):

        # Тут еще должна быть подготовка данных, типо заполнение пропусков
        self.data['alcohol'] = (5 * self.data['Dalc'] + 2 * self.data['Walc']) / 7
        self.data.drop_duplicates(keep = "first")

        self.features = self.data.drop(columns=["school", "sex", "address", "famsize", "Pstatus",
              "Mjob", "Fjob", "reason", "guardian", "schoolsup",
              "famsup", "paid", "activities", "nursery", "higher",
              "internet", "romantic", "Dalc", "Walc"])
        self.target = self.data['alcohol']

        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(
            self.features, self.target, test_size=0.2, random_state=17
        )

    def train_model(self):
        # Обучение какой-нибудь модели
        self.model = LinearRegression()
        self.model.fit(self.X_train, self.y_train)

    def predict(self, input_data):
        # Предсказание чего-нибудь от пользователя
        predictions = self.model.predict(input_data)
        return predictions

    def quality_model(self):
        # Оценка качества (хз, надо или нет) на тестовом куске
        y_pred = self.model.predict(self.X_test)
        mse = mean_squared_error(self.y_test, y_pred)
        print(f"MSE: {mse}")
```

```python
da = DatasetAnalyzer()

da.load_dataset()
da.prepare_data()

```
